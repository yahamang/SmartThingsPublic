```python
import FinanceDataReader as fdr
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
from datetime import datetime, timedelta
from itertools import combinations
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from sklearn.metrics import r2_score
```

```python
# 데이터 수집 및 전처리
day = 900

end_date = datetime.now() - timedelta(days=2)
start_date = end_date - timedelta(days=day)

# 비트코인 데이터
df_btc = fdr.DataReader('BTC/USD', start_date, end_date)
btc_price = df_btc['Close'] # 종가
btc_volume = df_btc['Volume'] # 거래량

# 금 가격
df_gold = fdr.DataReader('GC=F', start_date, end_date)
gold_price = df_gold['Close']

# 달러 인덱스
df_USD = fdr.DataReader('DX-Y.NYB', start_date, end_date)
USD = df_USD['Close']

# 나스닥
df_NASDAQ = fdr.DataReader('IXIC', start_date, end_date)
NASDAQ = df_NASDAQ['Close']

# 데이터 프레임 생성
df = pd.DataFrame({
    'Bitcoin_Price': btc_price,
    'Bitcoin_Volume': btc_volume,
    'Gold_Price': gold_price,
    'Dollar_Index': USD,
    'NASDAQ': NASDAQ
})

# 공백 매우기
df.fillna(method='ffill', inplace=True)
df.fillna(method='bfill', inplace=True)

# 입력 변수 조합 생성
xs = ['Bitcoin_Volume', 'Gold_Price', 'Dollar_Index', 'NASDAQ']
a = []
for i in range(len(xs)):
    b = list(combinations(xs, i + 1))
    for j in range(len(b)):
        a.append(list(b[j]))

best = 0
ys = []
Rs = []

# 각 조합에 대해 모델 학습 및 평가
for n in range(len(a)):
    print(a[n])
    dfx = df[a[n]]
    dfy = df[['Bitcoin_Price']] # 예측할 대상

    # 정규화
    scaler = MinMaxScaler()
    dfx_scaled = scaler.fit_transform(dfx)
    dfy_scaled = scaler.fit_transform(dfy)

    dfx_scaled = pd.DataFrame(dfx_scaled, index=dfx.index, columns=dfx.columns)
    dfy_scaled = pd.DataFrame(dfy_scaled, index=dfy.index, columns=dfy.columns)

    # list로 변환
    X = dfx_scaled.values.tolist()
    y = dfy_scaled.values.tolist()

    # 학습 사이즈
    window_size = 10

    # 학습용 데이터
    data_X = []
    data_y = []

    for i in range(len(y) - window_size):
        _X = X[i: i + window_size] # 다음 날 종가(i+windows_size)전까지
        _y = y[i + window_size]     # 다음 날 종가
        data_X.append(_X)
        data_y.append(_y)

    # train, test 분리 7:3
    train_size = int(len(data_y) * 0.7)
    train_X = np.array(data_X[0: train_size])
    train_y = np.array(data_y[0: train_size])

    test_size = len(data_y) - train_size
    test_X = np.array(data_X[train_size: len(data_X)])
    test_y = np.array(data_y[train_size: len(data_y)])

    # 학습 모델
    num = len(a[n]) # dfx 열 개수
    model = Sequential()
    model.add(LSTM(units=20, activation='relu', return_sequences=True, input_shape=(10, num)))
    model.add(Dropout(0.2))
    model.add(LSTM(units=20, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(units=1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    model.fit(train_X, train_y, epochs=70, batch_size=30) # 학습

    # 예측값
    pred_y = model.predict(test_X)
    ys.append(pred_y)

    # 평가
    R2 = r2_score(test_y, pred_y)
    Rs.append(R2)
    if R2 > best:
        best = R2
        best_a = a[n]

# 그래프 그리기
for n in range(len(a)):
    plt.figure()
    plt.plot(test_y, color='red', label='real BTC/USD')
    plt.plot(ys[n], color='blue', label='predicted BTC/USD')
    plt.title('BTC/USD prediction')
    plt.xlabel('time')
    plt.ylabel('BTC/USD')
    plt.legend()
    plt.show()
    print("조합 : ", a[n])
    print("R2 : ", Rs[n])
    print("내일 BTC :", df.Bitcoin_Price[-1] * ys[n][-1] / dfy_scaled.Bitcoin_Price[-1], 'USD')

print("가장 연관이 많은 자료 : ", best_a, "R2 : ", best)
```
